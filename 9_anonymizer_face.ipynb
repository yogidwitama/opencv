```python
import cv2
import mediapipe as mp
import os

img_path= "data/giga.jpeg"
img=cv2.imread(img_path)

H,W,_=img.shape

mp_face_detection=mp.solutions.face_detection

with mp_face_detection.FaceDetection(model_selection=0, min_detection_confidence=0.5) as face_detection:
    img_rgb=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    out= face_detection.process(img_rgb)
    
    if out.detections is not None:
        for detection in out.detections:
            location_data= detection.location_data
            bbox= location_data.relative_bounding_box
            
            x1,y1,w,h=bbox.xmin, bbox.ymin, bbox.width, bbox.height
            
            x1=int(x1 *W)
            y1=int(y1*H)
            w=int(w*W)
            h=int(h*H)
            
            # mask face
            img=cv2.rectangle(img,(x1,y1),(x1+w,y1+h),(255,0,0), 10)
            # bluring face
            # img[y1:y1+h, x1:x1+w,:]=cv2.blur(img[y1:y1+h,x1:x1+w,:],(30,30))
    cv2.imshow('img', img)
    cv2.waitKey(0)

# save image
# output_dir="data"
# cv2.imwrite(os.path.join(output_dir,"mask.png"), img)
```

## OOP Version

```python
import cv2
import mediapipe as mp
import os


def process_img(img, face_detection):
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    out = face_detection.process(img_rgb)
    
    if out.detections is not None:
        for detection in out.detections:
            location_data = detection.location_data
            bbox = location_data.relative_bounding_box
            
            x1, y1, w, h = bbox.xmin, bbox.ymin, bbox.width, bbox.height
            
            x1 = int(x1 * W)
            y1 = int(y1 * H)
            w = int(w * W)
            h = int(h * H)
            
            # Mask or blur the face
            img[y1:y1+h, x1:x1+w, :] = cv2.blur(img[y1:y1+h, x1:x1+w, :], (30, 30))
    
    return img


# Path file input
file_path = './data/giga.jpeg'

# Ambil direktori dari file_path
output_dir = os.path.dirname(file_path)

# MediaPipe face detection
mp_face_detection = mp.solutions.face_detection

with mp_face_detection.FaceDetection(model_selection=0, min_detection_confidence=0.5) as face_detection:
    img = cv2.imread(file_path)

    if img is not None:
        H, W, _ = img.shape
        img = process_img(img, face_detection)

        # Simpan hasil di direktori yang sama dengan input file
        output_path = os.path.join(output_dir, "mask.png")
        cv2.imwrite(output_path, img)
        print(f"Gambar berhasil disimpan di {output_path}")
    else:
        print(f"Error: Gambar tidak ditemukan di path: {file_path}")
```
